```python
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs, data = wavfile.read('./eval_samples/masna_piosenka.wav')
plt.plot(data)
plt.show()
```

```python
new_data = []
for it in range(len(data)):
    if 35000 < it < 45000:
        new_data.append(0.5*data[it])
    elif 55000 < it < 65000:
        new_data.append(0.7*data[it])    
    else:
        new_data.append(data[it])
        
new_data = np.array(new_data)
plt.plot(new_data)
plt.show()
```

```python
wavfile.write('./eval_samples/masna_piosenka_mod.wav', fs, new_data)
```
